# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python. It is classified as a micro-framework because it does not require particular tools or libraries and has no particular database or form handling libraries. It has a simple and easy-to-use syntax that makes it an ideal choice for small to medium-sized web applications.

### The advantages of Flask framework are:

1. Flexibility: Flask provides a lot of flexibility for developers to create a custom application structure as per their project requirements. Flask is highly adaptable, which means developers can customize the framework to their needs without the overhead of a full-stack framework like Django.

2. Lightweight: Flask is a lightweight framework that does not come with a lot of unnecessary features and tools. This means it is faster and more efficient than other full-stack frameworks.

3. Simplicity: Flask is simple and easy to learn. It has a small code base and intuitive syntax that allows developers to get started with the framework quickly.

4. Modular: Flask is highly modular, which means developers can add third-party libraries and extensions as needed. Flask has a lot of built-in support for popular extensions such as SQLALchemy, WTForms, and Flask-WTF.

5. Testing: Flask comes with a built-in testing framework that allows developers to easily test their applications. The framework supports unit testing, integration testing, and acceptance testing.

# Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is a mechanism that allows developers to map URLs to specific functions in the application. When a user requests a URL, Flask matches the URL to a function that can handle the request and returns the response to the user's browser.

In Flask, routing is achieved using the @app.route decorator. This decorator allows developers to define the URL route and the HTTP methods that the function should handle.

We use app routes in Flask to provide a clean and organized way to structure the application's URL endpoints. By defining app routes, we can separate the functionality of the application into different views, making it easier to maintain and extend the application.

App routes also make it easier to handle different types of HTTP requests, such as GET, POST, PUT, and DELETE. This is because each app route can be associated with a specific HTTP method, allowing the application to respond to different types of requests in different ways.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the Url_for() function.

In Flask, the url_for() function is used for URL building. This function generates a URL for a specific function in the application, based on the endpoint name of the function.
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return f'The URL for the about page is: {url_for("about")}'

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    app.run()

In this example, we have defined two routes: / and /about. The index() function returns a message that includes the URL for the about() function, which is generated using the url_for() function.

The url_for() function takes the endpoint name of the function as its argument and returns a URL for that endpoint. In this case, the endpoint name is "about", so the url_for() function returns the URL for the /about route.
